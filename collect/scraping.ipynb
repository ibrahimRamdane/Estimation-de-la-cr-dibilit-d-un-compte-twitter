```python
def get_last_tweets(name,count) :
    
    """_cette fonction retourne un dataframe pandas
    repr√©sentant le contenu des count-i√®mes derniers tweets de l'utilisateur name, ainsi que 
    le nombre de retweets, le nombre de likes et l'id de chacun des tweets_

    Returns:
        _tweets_df_: _dataframe pandas des objets sus-cit√©s_
    """
    
    api = twitter_setup()

    tweets = tweepy.Cursor(api.user_timeline, id = name, tweet_mode = 'extended').items(count)

    objects = ['Date','Id', 'Texte','RT','Nombre likes']

    tweets_list = [[tweet.created_at,  tweet.id,   tweet.full_text, tweet.retweet_count,tweet.favorite_count] for tweet in tweets]

    tweets_df = pd.DataFrame(tweets_list,columns=objects)
    
    print(tweets_df) #visu
    
    return tweets_df
```
